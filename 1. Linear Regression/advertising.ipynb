## Predicting sales based on advertising budget using linear regression.

```python
# Import dependencies
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split

# Supress warnings
import warnings
warnings.filterwarnings('ignore')
```

### Load Data

```python
# Load the dataset
advertising_dataset_path = 'data/advertising.csv'
# Replace dataset_path with the actual path to the dataset
df = pd.read_csv(advertising_dataset_path)
df 
```

### Data cleaning

```python
# Checking for null values
df.isnull().sum()
```

### There are no null values in the dataset

```python
# Outlier Analysis
df.plot.box()
plt.show()
```

## Plot the original data points

```python
# Create a list of labels for the features
feature_labels = ['TV', 'Radio', 'Newspaper']

# Define the size of the figure to be
fig = plt.figure(figsize=(10,6))

# Split the data into features (X) and target variable (y)
X = df[['TV', 'Radio', 'Newspaper']] 
y = df['Sales'] 

# Plot the original data points
plt.scatter(X['TV'], y, color='blue', label='TV')
plt.scatter(X['Radio'], y, color='green', label='Radio')
plt.scatter(X['Newspaper'], y, color='orange', label='Newspaper')
plt.xlabel('Advertising Budget')
plt.ylabel('Sales')
plt.title('Linear Regression - Advertising Budget vs. Sales')
plt.legend()
plt.show()
```

# Use scikit-learn to build a linear regression model

```python
# Create a linear regression model
model = LinearRegression()

# Fit the model to the data
model.fit(X, y)
```

### Note: In this initial case, the linear regression model is "trained" using all of the data.

```python
# Predict sales for new data points selected after visual inspection of the scatter plot for relevance
new_data = pd.DataFrame({'TV': [200, 400, 600],
                         'Radio': [40, 80, 120],
                         'Newspaper': [50, 100, 150]})
predicted_sales = model.predict(new_data)
```

```python
# Define figure size
fig = plt.figure(figsize=(10,6))

# Create a list of labels for the features
feature_labels = ['TV', 'Radio', 'Newspaper']

# Plot the original data points
plt.scatter(X['TV'], y, color='red', label='TV')
plt.scatter(X['Radio'], y, color='green', label='Radio')
plt.scatter(X['Newspaper'], y, color='orange', label='Newspaper')

# Plot the predicted sales
plt.scatter(new_data['TV'], predicted_sales, color='blue', label='Predicted Sales (TV)')
plt.scatter(new_data['Radio'], predicted_sales, color='purple', label='Predicted Sales (Radio)')
plt.scatter(new_data['Newspaper'], predicted_sales, color='yellow', label='Predicted Sales (Newspaper)')

# Add labels and title to the plot
plt.xlabel('Advertising Budget')
plt.ylabel('Sales')
plt.title('Sales Predictions for New Data: Linear Regression - Advertising Budget vs. Sales')
plt.legend()

# Display the plot
plt.show()
```

### Retrieve the coefficients for each of the 3 features: TV, Radio, Newspaper from model

```python
# Retrieve the coefficients and intercept
coefficients = model.coef_
# Retrieve the feature names
feature_names = X.columns.tolist()

# Create a dictionary of features and coefficients
feature_coefficients = dict(zip(feature_names, coefficients))
feature_coefficients
```

```python
feature_coefficients['TV']
```

```python
from pulp import LpProblem, LpVariable, LpMaximize

# Define the coefficients and max budget
c1, c2, c3 = feature_coefficients['TV'], feature_coefficients['Radio'], feature_coefficients['Newspaper']
Budget = max_budget

# Create the linear programming problem
model = LpProblem("Maximize Sales", LpMaximize)

# Define the decision variables
x1 = LpVariable("TV_budget", lowBound=0)
x2 = LpVariable("Radio_budget", lowBound=0)
x3 = LpVariable("Newspaper_budget", lowBound=0)

# Set the objective function
model += c1 * x1 + c2 * x2 + c3 * x3, "Maximize_Sales"

# Set the constraints
model += x1 + x2 + x3 <= Budget, "Budget_constraint"

# Solve the problem
model.solve()

# Print the optimal solution and sales
if model.status == 1:
    print("Optimal Solution:")
    print("TV Budget =", x1.varValue)
    print("Radio Budget =", x2.varValue)
    print("Newspaper Budget =", x3.varValue)
    print("Sales =", model.objective.value())
else:
    print("No feasible solution found.")
```

### Based on these inputs alone, the optimal solution to maximize sales would be to put every dollar toward Radio.

### Now set reasonable minimum departmental budgets using the mean and run model again.

```python
min_budget_tv = X.TV.mean()
min_budget_radio = X.Radio.mean()
min_budget_newspaper = X.Newspaper.mean()
```

```python
# Define the coefficients and max budget
c1, c2, c3 = feature_coefficients['TV'], feature_coefficients['Radio'], feature_coefficients['Newspaper']
Budget = max_budget

# Create the linear programming problem
model = LpProblem("Maximize Sales", LpMaximize)

# Define the decision variables
x1 = LpVariable("TV_budget", lowBound=0)
x2 = LpVariable("Radio_budget", lowBound=0)
x3 = LpVariable("Newspaper_budget", lowBound=0)

# Set the objective function
model += c1 * x1 + c2 * x2 + c3 * x3, "Maximize_Sales"

# Set the constraints
model += x1 >= min_budget_tv, "Minimum_TV_budget_constraint"
model += x2 >= min_budget_radio, "Minimum_Radio_budget_constraint"
model += x3 >= min_budget_newspaper, "Minimum_Newspaper_budget_constraint"
model += x1 + x2 + x3 <= Budget, "Budget_constraint"

# Solve the problem
model.solve()

# Print the optimal solution and sales
if model.status == 1:
    print("Optimal Solution:")
    print("TV Budget =", x1.varValue)
    print("Radio Budget =", x2.varValue)
    print("Newspaper Budget =", x3.varValue)
    print("Sales =", model.objective.value())
else:
    print("No feasible solution found.")
```

```python
# Create a list of labels for the features
feature_labels = ['TV', 'Radio', 'Newspaper']

# Define the size of the figure to be
fig = plt.figure(figsize=(10,6))

# Split the data into features (X) and target variable (y)
X = df[['TV', 'Radio', 'Newspaper']] 
y = df['Sales'] 

# Plot the original data points
plt.scatter(X['TV'], y, color='red', label='TV')
plt.scatter(X['Radio'], y, color='green', label='Radio')
plt.scatter(X['Newspaper'], y, color='orange', label='Newspaper')

# Retrieve the optimal solution values
optimal_tv_budget = x1.varValue
optimal_radio_budget = x2.varValue
optimal_newspaper_budget = x3.varValue
optimal_sales = model.objective.value()

# Plot the optimal solution points
plt.scatter(optimal_tv_budget, optimal_sales, color='blue', edgecolor='black', label='Optimal Solution')

plt.xlabel('Advertising Budget')
plt.ylabel('Sales')
plt.title('Optimal Sales plotted against Historical Advertising Budgets vs. Sales')
plt.legend()
plt.show()
```

# Next practice visualizing data in 3 dimensions

```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
import matplotlib.cm as cm

# Generate random advertising data
np.random.seed(1)
n_samples = 100
TV_budget = np.random.uniform(low=0, high=500, size=n_samples)
radio_budget = np.random.uniform(low=0, high=100, size=n_samples)
newspaper_budget = np.random.uniform(low=0, high=150, size=n_samples)
sales = 10 + 0.6 * TV_budget + 0.3 * radio_budget + 0.1 * newspaper_budget + np.random.normal(0, 5, size=n_samples)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# Plot the data points
ax.scatter(TV_budget, radio_budget, newspaper_budget, c=sales, cmap='viridis')

# Set labels for the axes
ax.set_xlabel('TV Budget')
ax.set_ylabel('Radio Budget')
ax.set_zlabel('Newspaper Budget')

# Set title for the plot
ax.set_title('Randomly Generated Data: Advertising Budget vs. Sales')

# Add a colorbar
cbar = plt.colorbar(mappable=cm.ScalarMappable(cmap='viridis'))
cbar.set_label('Sales')

# Show the plot
plt.show()
```
